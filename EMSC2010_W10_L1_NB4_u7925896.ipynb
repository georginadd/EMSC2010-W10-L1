````python
import numpy as np # NumPy for numerical computing
import matplotlib.pyplot as plt #plotting library
import pandas as pd #Pandas for data handling
````

````python
blossoms = 'blossom_doy.xlsx'
df = pd.read_excel(blossoms)
df.head()
````

````python
plt.plot(df.year,df.doy, 'k') #plot the data
plt.xlabel('Year') #label the x-axis
plt.ylabel('Day of year') #label the y-axis
plt.minorticks_on() #add minorticks to the plot
````

````python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
````

````python
w = 51 # length of the moving average
doy_s = moving_average(df.doy, w) # process the signal
year_s = moving_average(df.year, w) # process the times
````

````python
plt.plot(df.year,df.doy, '.k') #plot the data
plt.plot(year_s,doy_s,'r', label="Moving average")
plt.legend()

plt.xlabel('Year')
plt.ylabel('Day of the first bloom')
plt.minorticks_on()
````

We can see that from around 1850, the day the cherry blossoms appear has moved earlier in the year. It's getting warmer, spring is happening earlier, earlier blooming trend
